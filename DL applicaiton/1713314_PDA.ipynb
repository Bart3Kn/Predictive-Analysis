```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import keras #helps for ANN model construction
import tensorflow as tf #helps for ANN model construction
from sklearn.preprocessing import StandardScaler #helps for standardation of input data 
```

# Loading the data
I am going to be using the preprocessed data from the previous section, I do not need to perform and EDA on the data here and do no need to clean/process the data, as this has already been done.

```python
df = pd.read_csv('/content/drive/MyDrive/PDAcw/data.csv')

#Checking if the data imported in correctly
print(df.columns)
print(df.shape)
print(df.head())
print(df.info())
```

Due to the issue of tensorflow not being able to work with int64, I am going to have to convert a few columns to a float64 data type. Furthermore, I also need to convert death event to a Boolean as that should be the correct data type.

```python
df['creatinine_phosphokinase'] = df['creatinine_phosphokinase'].astype('float')
df['ejection_fraction'] = df['ejection_fraction'].astype('float')
df['serum_sodium'] = df['serum_sodium'].astype('float')
df['time'] = df['time'].astype('float')
df['DEATH_EVENT'] = df['DEATH_EVENT'].astype('bool')

print(df.info())
print(df.head())
```

## Splitting the data
I am going to split the data in the same 70/30 split for train/test as I did for machine learning, this is to ensure that I am able to compare the outcome of the results much more accurately with one another.

```python
X = df.drop(['DEATH_EVENT','anaemiaCH','diabetesCH','high_blood_pressureCH','smokingCH','sexCH'], axis=1)
Y = df['DEATH_EVENT']

#Using a scaler to help with the performance of the model as it was having problems with processing the data input type.
#This sample code seems to have resolved my issue.
colnames = list(X.columns)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = colnames)
X_scaled.describe().T

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=63)
```

# Building and Training the Model
In order for me to successly fully create a deep learning model, I will need to create and train one on the training data. As such, I will need to construct the shape and size. I will be using keras for this



```python
#Model creation toolkit
model = keras.Sequential()

#Initial Layer with input nodes
model.add(keras.layers.Dense(10, input_shape=(13,))) #Needs to be adjusted to correct size
model.add(keras.layers.Activation("relu"))
#Layer 2
model.add(keras.layers.Dense(10))
model.add(keras.layers.Activation("relu"))
#Layer 3
model.add(keras.layers.Dense(10))
model.add(keras.layers.Activation("relu"))
#Output
model.add(keras.layers.Dense(1)) # 1 output for DEATH_EVENT
#Creation on 
model.compile(optimizer='SGD', loss = 'binary_crossentropy', metrics=['accuracy'])


#Fitting the model to the data is causing issues
history = model.fit(X_train, Y_train, epochs = 300, batch_size = 64,verbose = 1)
```

0.8+ accuracy can be problematic as it may not be the most accurate prediction model for heart failure, but it may be able to produce enough accurate results to warrant significant accuracy.

# Predicting on Test set
I am now going to use the trained model on the test set to see how it performs

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

Y_pred = model.predict(X_test)
Y_pred = (Y_pred>0.5)

matrix = confusion_matrix(Y_test, Y_pred)
tn = matrix[0,0]
fp = matrix[0,1]
fn = matrix[1,0]
tp = matrix[1,1]


print("True Negative: ", tn)
print("False Positive: ",  fp)
print("False Negative: ",  fn)
print("True Negative: ",  tp)


accuracy = ((tn+tp)/(tn + fp+ fn +tp))*100
print("Accuracy on the test dataset is:", accuracy, "%")

print("\n\nTesting Results:\n")
print(classification_report(Y_test, Y_pred))

```
